```python
import pandas as pd
data=pd.read_csv("C:/Users/Lenovo/Downloads/synthetic_salary_data_raw.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
x_feature = 'YearsExperience'   
y_feature = 'Salary'            

plt.figure(figsize=(8, 6))
plt.scatter(data[x_feature], data[y_feature], color='blue', alpha=0.6)
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.title(f'Scatter Plot: {x_feature} vs {y_feature}')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
x=data['YearsExperience']
y=data['Salary']
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(model.intercept_)
print(model.coef_[0])
```
